# Fuel price tracking

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data_url = 'https://github.com/jbjbjb1/demo_fuel_project/blob/master/fuel_data.pkl?raw=true'
df = pd.read_pickle(data_url)
df.head()
```

```
ax = df.plot(x="Date", y="Fuel_price", legend=False)
ax2 = ax.twinx()
df.plot(x="Date", y="Brent_price_aud", ax=ax2, legend=False, color="r")
ax.figure.legend()
plt.show()
```
